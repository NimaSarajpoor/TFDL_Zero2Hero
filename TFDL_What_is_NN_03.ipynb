# 4. format of X and y in NN?

# 5. Logisitic Regression?

**NOTE:** This is NOT about doing logistic regression in NN. It rather shows the concept of logistic regression, and we will use it later in NN context!

$x$: input (with $n$ features) --> $\hat{y} = P(y=1 | x)$

and, in logistic regression, we have this: <br>

$\hat{y} = \frac{1}{1 + e ^ {-z}}$, where $z = w^{T}x + b$.

So, we can say $\hat{y} = \sigma(w^{T}x + b)$

So, the parameters are: $\theta = (\theta_{0}, \theta_{1}, ..., \theta_{n}) ^ {T}$. Note that the $\theta_{0}$ corresposnds to $x_{0} = 1$. So, our input x becomes: <br>

$x' = (x_{0}, x_{1}, ..., x_{n}) ^ {T}$, and thus: $\hat{y} = \sigma(\theta^{T}x')$

**NOTE:** Now that we define our model for predicting $y$, how can we find the parameters $w$ and $b$. We need to find $w$ and $b$ such that it minimizes the error between true $y$ and its estimation $\hat{y}$ on all observations. So, we need to know two things:
* How to combine error of all observations?
* What is error? (a.k.a loss)

We start with the second question! The error in logistic regression can be defined as:

$
L(\hat{y}, y) = - \left(
ylog(\hat{y}) + (1-y)log(1-\hat{y}
\right)
$

Now, think about two cases `y=0` and `y=1`... When `y=0`, the loss becomes $-log(1-\hat{y})$. To lower the loss, the problem tries to increase $log(1-\hat{y})$. Therefore, it tries to reduce $\hat{y}$. (Note that according to the formula provided for $\hat{y}$ in logistic regression, the $\hat{y}$ changes from 0 to 1.) So, when `y=0`, problem tries to make $\hat{y}$ equal 0 as well! Cool! 

<br>

**Exercise:** Think about case when `y=1`.

So, this loss function work. 

<br>

**NOTE:** Why not something like $(y - \hat{y})^{2}$? Because it creates ugly function that is not good in optimization!)

Now, the first question was how to combine the errors of all observations? The answer is simple here: Just add them! So, the **cost function** becomes:

$\frac{1}{m} \sum\limits_{i=1}^{m} L\left(
\hat{y(i)}, y(i)
\right)$

Now, the goal is to find $w$ and $b$ that minimizes the above cost function.

## 5.1 Gradient Descent

Before talking about optimizing the aforementioend cost function (and find optimal $w$ and $b$), we need to talk about Gradient Descent!

Go through the following resources (preferably in the same order presented below).

[Khan Academy: Gradient series](https://www.khanacademy.org/math/multivariable-calculus/multivariable-derivatives/gradient-and-directional-derivatives/v/gradient)

<br>

[Khan Academy: Gradient Descent](https://www.khanacademy.org/math/multivariable-calculus/applications-of-multivariable-derivatives/optimizing-multivariable-functions/a/what-is-gradient-descent)


Let's say we have parameter vector $\theta$... How should we change it to get lower error? Note that we have a vector $\theta$. So, the question is, what would be that vector? In fact, the question can be to what direction and to what extent?!

<br>

We learned that the direction that maximize $f(\theta)$ is $\frac{\nabla f(\theta)}{||\nabla f(\theta)||}$. So, the direction that minimize f is the negative of such direction!
But, what about the step we should take in such direction? Well, that is a parameter that is called learning rate. Let us call this parameter $\alpha$. So, to update $\theta$, we may have: 

<br>

$\theta = \theta + \alpha (- \frac{\nabla (f)}{||\nabla (f)||})$

<br>


But, we should better write it as: 

$\theta = \theta + \alpha \left(- \nabla f(\theta) \right)$

Why? Because when the steep is large, we can have bigger change, and when the steep is small (i.e. we are probably close to the optimal point), we will have smaller changes.



**Now, let's get back to our problem.** We have a cost function (as a function of $w$ and $b$) and we want to minimize it. How should we change those parameters to move into the optimal point?

To update vector $\theta$, we can do:

$\theta = \theta - \alpha (d\theta)$ 

where, $d\theta$ is the directional derivitve of cost function at point $\theta$.

**NOTE:** <br>

If we consider $w$ and $b$ seperately, it becomes:

$w = w - \alpha dw$, where $dw$ is $\frac{\partial J(w, b)}{\partial w}$
<br>
$b = w - \alpha db$, where $db$ is $\frac{\partial J(w, b)}{\partial b}$

<br>
$J(.)$ is the cost function formulated earlier.

Before we move forward, one may ask: derivative of $J$ with respect to what variable? Recall that the total cost $J$ is a function of many variables ($W$ and $b$) and to find a (local) optimal point that minimizes $J$, we should move in the opposite direction of Gradient Ascent. The directional derivative is constructed by using the partial derivative of $J$ with respect to each of those variables. 

Now, let us take a look at the computation graph above. This is just to practic the concpt==The goal is to take the (partial) derivative of $J$ with respect to $v$, $u$, $a$, $b$, and $c$.



$$
\begin{equation}
\frac{d(J)}{d(v)} = 3
\end{equation}
$$


But, note that we want to find the gradient of J, which contains partial derivatives of J with respect to the variables. So, let us find: $\frac{\partial(J)}{\partial(a)}$


$$
\begin{equation}
\frac{\partial(J)}{\partial(a)} = \frac{\partial(J)}{\partial(V)} \times \frac{\partial(V)}{\partial(a)}
\end{equation} \quad {ChainRule}
$$ 


We can do the same for calculating partial derivative of $J$ with respect to other variables. For instance:


$$
\begin{equation}
\frac{\partial(J)}{\partial(c)} = \frac{\partial(J)}{\partial(V)} \times \frac{\partial(V)}{\partial(U)} \times \frac{\partial(U)}{\partial(c)}
\end{equation} \quad {ChainRule}
$$ 


### 5.2.1: Example on Logistic  Regression 

```python
Image(filename = PATH + "logistic_regression_graph.png", width=500, height=500)
```

The graph (except the last box) is a simple logistic regression with two variables (features). The goal is to minimize the loss function $L$. Recall that the $x_{1}$ and $x_{2}$ are from data and thus, they are NOT variables. The variables are $w_{1}$, $w_{2}$, and $b$. We would like to find the optimal decision variables that minimize loss function $L$.

Also, recall that to minimize $L$, we need to move in the opposite direction of Gradient Ascent. Hence, the method we use is called "Gradient Descent!" :) 

How should we update variables so that we move toward the optimal point?
$var := var - \alpha \times \frac{\partial(obj\_function)}{\partial(var)} $

So, we need to find the partial derivative of $L$ with respect to each of the variables $w_{1}$, $w_{2}$, and $b$. Let's do it! We start from right to left

$L(\hat{y}, y) = -ylog(\hat{y}) - (1 - y)log(1-\hat{y})$
<br>
$\frac{\partial(L)}{\partial(\hat{y})} = \frac{(-y)(1)}{\hat{y}} -  \frac{(1-y)(-1)}{1-\hat{y}} = \frac{-y}{\hat{y}} + \frac{(1-y)}{1-\hat{y}}$

NOTE: are you thinking about the base in log? It does not matter. In calculation above, I assumed `log` is `natural log`, i.e. `ln`. However, if I consider base 10 or 2, I will end up with the same relationship provided above but just containing a new factor $\frac{1}{ln(base)}$

I do not care about factor, because its impact can be merged with the impact of learning rate!

The sigmoid function $\sigma$ is defined as: $\hat{y} = \sigma(z) = \frac{1}{1 + e^{-z}}$

To get to the variables, I need to go further and calculate the drivative of $L$ with respsect to $z$


$$
\begin{align}
\frac{\partial(L)}{\partial(z)} ={}& 
\frac{\partial(L)}{\partial(\hat{y})} \times \frac{\partial(\hat{y})}{\partial(z)} \\
={}& \left(\frac{-y}{\hat{y}} + \frac{(1-y)}{1-\hat{y}}\right) \times \left(\hat{y} \times (1 - \hat{y})\right) \\
={}& \hat{y} - y
\end{align}
$$

But, what is $z$? It is an intermediate  variable is calculated as follows:

$z = w_{1} \cdot x_{1} + w_{2} \cdot x_{2} + b$

Now, we are at the position to calculate $L$ with respect to the variables $w_{1}$, $w_{2}$, and $b$...


$$
\begin{align}
\frac{\partial(L)}{\partial(w_{1})} ={}& 
\frac{\partial(L)}{\partial(z)} \times \frac{\partial(z)}{\partial(w_{1})} \\
={}& (\hat{y} - y) \times x_{1}
\end{align}
$$


Alternatively, we can say: <br> $dw_{1} = x1 \cdot dz $

In other words, after calculating $dz$ (the deriviative  of $L$ w.r.t variable z, which is linear combination of input), we can calculate $dw_{1}$, $dw_{2}$, and $b$. Now that we have them, we can update them as follows:
<br>
$w_{1} := w_{1} - \alpha \cdot dw_{1}$
<br>
and same for others.

Note that we just discussed the backward propogation for single input. What if we have mulitple inputs? For instance, with $m$ inputs (each with $n$, say $n=2$, features), we will get $m$ different values for $dw_{1}$, and other variables. But, we need a single value as we need to make decision on how much we should update $w_{1}$ (via $w_{1} = w_{1} - \alpha \cdot dw_{1}$). We can take that single value by taking average of $dw$s calculated for the $m$ inputs. So:
<br>

$dw_{1} = \frac{1}{m} \sum\limits_{i=1}^{m} dw^{(i)}_{1}$, where $dw^{(i)}_{1}$ is the derivative of loss function with respect to $w_{1}$ when input is $x_{i}$.

**Note:** We should often vectorize our calculation (and take advantage of Numpy, Numba, etc) to speed up our calculation.

**EXAMPLE:**<br>
Let's say we have three observations $x^{(1)}$, $x^{(2)}$, and $x^{(3)}$. 

A non-vectorized approach goes like this: <br>
$z^{(1)} = w^{T}x^{(1)} + b$ <br>
$z^{(2)} = w^{T}x^{(2)} + b$ <br>
$z^{(3)} = w^{T}x^{(3)} + b$ <br>

Note that we are using the same variables $W$ and $b$. Not that we cannot have more, but this is just an example of part of neuron that takes input $x$ and calculate $z$ (and then pass it to an acitvation function, like sigmoid $\sigma$). 

Now, we can vectorize it as follows: <br>
$[z^{(1)}, z^{(2)}, ..., z^{(m)}] = w^{T}[x^{(1)}, x^{(2)}, ..., x^{(m)}] + b$ <br>

`# in python`
Z = np.dot(w, X) + b

Note that the `i`-th column of Z (and X) correponds the `i`-th observation.

```python
m_samples = 3
n_feature = 2

seed = 100
np.random.seed(seed)

W = np.random.randint(low=0, high=9, size=n_feature)
print('W: \n', W)

X = np.random.randint(low=0, high=9, size=(n_feature, m_samples))
print('X: \n', X)
```

```python
np.dot(W, X) 
# this is  sum product between W and second-to-last axis of X ... 
# so, W[0] will be multiplied to row 0 of X (the 0-th feature), 
# W[1] will be multiplied to row 1 of X (the 1-th feature),
# and so on...
```

```python
# for input matrix X=[X(1), X(2), ..., X(m)].
# NOTE: X(i) is a column vector of lenth n, containing n feature value of 
# the i-th observation

def _update_parameters(X, y, w, b, learning_rate=0.001):
    """
    Update W and b via Gradient Descent
    
    Parameters
    ----------
    X: numpy.ndarray
        The feature matrix, n-by-m, where `n` is is number of features,
        and `m` is number of samples. 
    
    y: numpy.ndarray
        The target values. A 1D array whose length must be equal to 
        the number of samples `m`
    
    W: numpy.ndarray
        The weights (in z = w^{T}X + b), a 1D array of length `n`. `W[i]` is
        the weight corresponds to i-th feature.
    
    b: float
        The bias (in z = w^{T}X + b)
    
    learning_rate: float
        The step in updating decision variables in gradient descent
    
    Returns:
    --------
    None
    """
    z = np.matmul(w.reshape(1,-1), X) + b 
    # z is 1d array
    # b is scalar, but will be broadcasted to 1D array
    
    y_pred = 1 / (1 + np.exp(-z))
    dz = y_pred - y
    
    dw = np.mean(np.matmul(X, dz.T))
    db = np.mean(dz)
    
    w -= learning_rate * dw
    b -= learning_rate * db
    
# source: https://www.coursera.org/learn/neural-networks-deep-learning/lecture/IgFnJ/vectorizing-logistic-regressions-gradient-output
```

**NOTE:**

```python
#example:
seed = 0
np.random.seed(seed)

X = np.random.rand(10, 100) #10 rows as 10 features, and 100 as 100 observations 
# (recall that we now put each observation in one column)
y = np.random.rand(100) # y values!

w = np.zeros(10, dtype=np.float64)
b = 0
_update_parameters(X, y, w, b)

print('updated w:', w)
print('updated b:', b)
```
