# 2. What does Neual Network (NN) look like?

It is a bunch of layers (each layer with bunch of codes) and they are connected to each other...

```python
Image(filename = PATH + "NN_simple.jpg", width=300, height=300)
# source: https://stackoverflow.com/questions/35345191/what-is-a-layer-in-a-neural-network
```

Here we can see our NN has four layers. Input, output, and two hidden layers. (**NOTE:** Sometimes you may see the input layer does not appear in a picture and there is just some $x_{1}$, ..., $x_{m}$ connected to the nodes in the first layer.)

This is an standard NN. There are more advanced ones like CNN, RNN, .... BUT, do not worry and do not think about them. It will be covered later. That was just a heads up! :D

**NOTE:** A very good application of NN (and Deep Learning) is on understanding underlying structure of "unstructured data"! In tabuldar data, we often have clear features for each individual. However, in unstructured data (e.g. voice of a person), there is clear feature. Other examples, can be an image of a flower, or a text in Google Reivew. All of these have no columns of feature. NN seems to do a  very good job in understanding them! 

So, the lesson that we learn from this is: <br>
NN can be applied to structured data. But, it is really power full in performing classification (and/or other ML tasks) on images/texts/audio.
