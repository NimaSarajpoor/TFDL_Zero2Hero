# 5. Logisitic Regression?

**NOTE:** This is NOT about doing logistic regression in NN. It rather shows the concept of logistic regression, and we will use it later in NN context!

$x$: input (with $n$ features) --> $\hat{y} = P(y=1 | x)$

and, in logistic regression, we have this: <br>

$\hat{y} = \frac{1}{1 + e ^ {-z}}$, where $z = w^{T}x + b$.

So, we can say $\hat{y} = \sigma(w^{T}x + b)$

So, the parameters are: $\theta = (\theta_{0}, \theta_{1}, ..., \theta_{n}) ^ {T}$. Note that the $\theta_{0}$ corresposnds to $x_{0} = 1$. So, our input x becomes: <br>

$x' = (x_{0}, x_{1}, ..., x_{n}) ^ {T}$, and thus: $\hat{y} = \sigma(\theta^{T}x')$

**NOTE:** Now that we define our model for predicting $y$, how can we find the parameters $w$ and $b$. We need to find $w$ and $b$ such that it minimizes the error between true $y$ and its estimation $\hat{y}$ on all observations. So, we need to know two things:
* How to combine error of all observations?
* What is error? (a.k.a loss)

We start with the second question! The error in logistic regression can be defined as:

$
L(\hat{y}, y) = - \left(
ylog(\hat{y}) + (1-y)log(1-\hat{y}
\right)
$

Now, think about two cases `y=0` and `y=1`... When `y=0`, the loss becomes $-log(1-\hat{y})$. To lower the loss, the problem tries to increase $log(1-\hat{y})$. Therefore, it tries to reduce $\hat{y}$. (Note that according to the formula provided for $\hat{y}$ in logistic regression, the $\hat{y}$ changes from 0 to 1.) So, when `y=0`, problem tries to make $\hat{y}$ equal 0 as well! Cool! 

<br>

**Exercise:** Think about case when `y=1`.

So, this loss function work. 

<br>

**NOTE:** Why not something like $(y - \hat{y})^{2}$? Because it creates ugly function that is not good in optimization!)

Now, the first question was how to combine the errors of all observations? The answer is simple here: Just add them! So, the **cost function** becomes:

$\frac{1}{m} \sum\limits_{i=1}^{m} L\left(
\hat{y(i)}, y(i)
\right)$

Now, the goal is to find $w$ and $b$ that minimizes the above cost function.

## 5.1 Gradient Descent

Before talking about optimizing the aforementioend cost function (and find optimal $w$ and $b$), we need to talk about Gradient Descent!

Go through the following resources (preferably in the same order presented below).

[Khan Academy: Gradient series](https://www.khanacademy.org/math/multivariable-calculus/multivariable-derivatives/gradient-and-directional-derivatives/v/gradient)

<br>

[Khan Academy: Gradient Descent](https://www.khanacademy.org/math/multivariable-calculus/applications-of-multivariable-derivatives/optimizing-multivariable-functions/a/what-is-gradient-descent)


Let's say we have parameter vector $\theta$... How should we change it to get lower error? Note that we have a vector $\theta$. So, the question is, what would be that vector? In fact, the question can be to what direction and to what extent?!

<br>

We learned that the direction that maximize $f(\theta)$ is $\frac{\nabla f(\theta)}{||\nabla f(\theta)||}$. So, the direction that minimize f is the negative of such direction!
But, what about the step we should take in such direction? Well, that is a parameter that is called learning rate. Let us call this parameter $\alpha$. So, to update $\theta$, we may have: 

<br>

$\theta = \theta + \alpha (- \frac{\nabla (f)}{||\nabla (f)||})$

<br>


But, we should better write it as: 

$\theta = \theta + \alpha \left(- \nabla f(\theta) \right)$

Why? Because when the steep is large, we can have bigger change, and when the steep is small (i.e. we are probably close to the optimal point), we will have smaller changes.



**Now, let's get back to our problem.** We have a cost function (as a function of $w$ and $b$) and we want to minimize it. How should we change those parameters to move into the optimal point?

To update vector $\theta$, we can do:

$\theta = \theta - \alpha (d\theta)$ 

where, $d\theta$ is the directional derivitve of cost function at point $\theta$.

**NOTE:** <br>

If we consider $w$ and $b$ seperately, it becomes:

$w = w - \alpha dw$, where $dw$ is $\frac{\partial J(w, b)}{\partial w}$
<br>
$b = w - \alpha db$, where $db$ is $\frac{\partial J(w, b)}{\partial b}$

<br>
$J(.)$ is the cost function formulated earlier.

Before we move forward, one may ask: derivative of $J$ with respect to what variable? Recall that the total cost $J$ is a function of many variables ($W$ and $b$) and to find a (local) optimal point that minimizes $J$, we should move in the opposite direction of Gradient Ascent. The directional derivative is constructed by using the partial derivative of $J$ with respect to each of those variables. 

Now, let us take a look at the computation graph above. This is just to practic the concpt==The goal is to take the (partial) derivative of $J$ with respect to $v$, $u$, $a$, $b$, and $c$.



$$
\begin{equation}
\frac{d(J)}{d(v)} = 3
\end{equation}
$$


But, note that we want to find the gradient of J, which contains partial derivatives of J with respect to the variables. So, let us find: $\frac{\partial(J)}{\partial(a)}$


$$
\begin{equation}
\frac{\partial(J)}{\partial(a)} = \frac{\partial(J)}{\partial(V)} \times \frac{\partial(V)}{\partial(a)}
\end{equation} \quad {ChainRule}
$$ 


We can do the same for calculating partial derivative of $J$ with respect to other variables. For instance:


$$
\begin{equation}
\frac{\partial(J)}{\partial(c)} = \frac{\partial(J)}{\partial(V)} \times \frac{\partial(V)}{\partial(U)} \times \frac{\partial(U)}{\partial(c)}
\end{equation} \quad {ChainRule}
$$
